```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/creditcard.csv")
```

```python
corr_matrix = df.corr() # returns a matrix with the correlation with each feature and other features
# print(corr_matrix)

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, vmin=-1, vmax=1)
plt.title('Feature Correlation Heatmap')
```

```python

class_correlations = df.corr()[['Class']].sort_values(by='Class', ascending=False)

#Draw a vertical strip heatmap
plt.figure(figsize=(4, 10))

sns.heatmap(class_correlations, vmin=-1, vmax=1, annot=True, cmap='coolwarm')

plt.title('Feature Correlation with Fraud')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
X = df.drop("Class", axis = 1)
y = df["Class"]

model = RandomForestClassifier(random_state= 42, n_jobs = -1)
model.fit(X, y)

scores = pd.Series(model.feature_importances_, index = X.columns)
scores
```

```python
scores.sort_values(ascending= False).plot(kind="bar", figsize=(12, 6))
plt.title("Feature Correlation: Bar Graph")
```
